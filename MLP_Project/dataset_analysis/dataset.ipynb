### Dataset 불러오기

order_data: 소비자의 주문 내역, 한 유저당 이 사이트에서 구입한 order의 개수 (Order의 상세 내용은 order_products에 존재) <br>
prior_data: 한 Order_id에 들어있는 Product들의 데이터 <br>
Product_data: 한 Product에 대해서 Department와 Aisle에 대한 정보 <br>
Aisle_data: Aisle 번호에 대한 이름 <br>

```python
order_data = pd.read_csv("dataset/orders.csv", encoding='UTF8')
prior_data = pd.read_csv("dataset/order_products__prior.csv", encoding='UTF8')
product_data = pd.read_csv("dataset/products.csv", encoding='UTF8')
aisle_data = pd.read_csv("dataset/aisles.csv", encoding='UTF8')
```

```python
df1 = pd.merge(prior_data,product_data,on='product_id')
df1.head(5)
df1 = pd.merge(df1,order_data,on='order_id')
df_final = pd.merge(df1,aisle_data, on='aisle_id')
df_final = df_final.sort_values(by=['user_id','order_id'])
```

```python
df_final = df_final.loc[:,["order_id","user_id","product_id","aisle_id"]]
for i in range(134):
    print(i+1,":",len(df_final[df_final["aisle_id"] == i+1]))
# df_final.to_csv("dataset/data_final.csv", index=False)
```
